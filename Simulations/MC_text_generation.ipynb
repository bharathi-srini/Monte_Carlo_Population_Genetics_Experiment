```python
import numpy as np
```

```python
folder = '/Users/BharathiSrinivasan/Documents/HU-MEMS-Sem4/NIS/NIS-Monte_Carlo/Simulations/trump-speeches/'
speeches = open(folder+'speeches.txt', encoding='utf8').read()
```

```python
corpus = speeches.split()
```

```python
corpus
```

```python
def make_pairs(corpus):
    for i in range(len(corpus)-1):
        yield (corpus[i], corpus[i+1])
```

```python
pairs = make_pairs(corpus)
```

```python
word_dict = {}
```

```python
for word_1, word_2 in pairs:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

```python
first_word = np.random.choice(corpus)
```

```python
while first_word.islower():
    first_word = np.random.choice(corpus)
```

```python
chain = [first_word]
```

```python
#Set length of sentence to generate
n_words = 100
```

```python
for i in range(n_words):
    chain.append(np.random.choice(word_dict[chain[-1]]))
```

```python
print(' '.join(chain))
```
